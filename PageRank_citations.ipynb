```python
import pandas as pd
import networkx as nx
df=pd.read_csv(r'C:\Users\quilo\Desktop\2020-08-20T18_12_28_2.csv')
citing=list(df['citing'])
cited=list(df['cited'])
```

```python
# Criar grafo dirigido vazio.
G = nx.DiGraph()
for i in  range(len(citing)):
    G.add_edge(citing[i],cited[i])
```

```python
PageRank = nx.pagerank(G)
```

```python
PageRank_sorted={k: v for k, v in sorted(PageRank.items(), key=lambda item: item[1],reverse=True)}
PageRank_sorted
```
